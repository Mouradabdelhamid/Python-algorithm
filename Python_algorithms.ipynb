```python
def binary_search(sequence, item):  
  begin_index = 0 
  end_index = len(sequence) - 1
  mid = 0

  while begin_index <= end_index:
    mid = (end_index + begin_index)//2 
    if sequence[mid] == item:
      return mid
    elif sequence[mid] > item:
      end_index = mid - 1
    else:
      begin_index = mid + 1
  return 0 
list_1 = [1,2,3,5,8]
x = 6
result = binary_search(list_1, x)
if result == 0:
  print('The item is not found.')
else:
  print('The element is in the list at index ' + str(result))
```

```python
def power (a, b):
  if b == 0:
    return 1
  elif b == 1:
    return a
  else:
    return a**b
print(power(3, 4))
```

```python
def bubble_sort(list_1):
  for n in range(len(list_1) -1, 0, -1):  # looping from last index until first index
    for i in range(n):                    
      if list_1[i] > list_1[i + 1]:       
        temp = list_1[i]                 
        list_1[i] = list_1[i+1]          #swapping
        list_1[i+1] = temp               #swapping
  return list_1                          #ordered list
list_1 = [29,13,22,37,52,49,46,71,56]
bubble_sort(list_1)
```

```python
def merge_sort(list_1):
  if len(list_1) > 1:
    mid = len(list_1) // 2
    left_list = list_1[:mid]
    right_list = list_1 [mid:]

    merge_sort(left_list)
    merge_sort(right_list)

    i = 0
    j = 0
    k = 0

    while i < len(left_list) and j < len(right_list):
      if left_list[i] < right_list[j]:
        list_1[k] = left_list[i]
        i += 1
        k += 1
      else:
        list_1[k] = right_list[j]
        j += 1
        k += 1
    while i < len(left_list):
      list_1[k] = left_list[i]
      i += 1
      k += 1
    while j < len(right_list):
      j += 1
      k += 1
list_1 = [29,13,22,37,52,49,46,71,56]
merge_sort(list_1)
print(list_1)
```

```python
def partition(my_list, min, max):
    pivot = my_list[max]
    i = min - 1
    for j in range(min, max):
        if my_list[j] <= pivot:
            i = i + 1
            (my_list[i], my_list[j]) = (my_list[j], my_list[i])
    (my_list[i + 1], my_list[max]) = (my_list[max], my_list[i + 1])
    return i + 1


def quickSort(my_list, min, max):
    if min < max:
        sub_array = partition(my_list, min, max)
        quickSort(my_list, min, sub_array - 1)
        quickSort(my_list, sub_array + 1, max)
        
my_list =[29,13,22,37,52,49,46,71,56]
size = len(my_list)
quickSort(my_list, 0, size - 1)
print(my_list)
```
